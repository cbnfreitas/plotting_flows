```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
X, T = np.mgrid[-0.5:1.5:0.01, -1:1:0.01]
DT = 1 + 0*X  #This 0*X is only a trick to make DT with the proper size
DX = X * (1 - X)

#Plotting arrows 
plt.streamplot(T, X, DT, DX, density=0.8)

#Plotting equilibrium points
t = T[0]  #It sets t as -1:1:0.01
plt.plot(t, 1 + 0*t, 'black')
plt.plot(t, 0 + 0*t, 'black')

plt.xlabel('t')
plt.ylabel('x')
```
